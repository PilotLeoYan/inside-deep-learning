# Dataset

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

X = housing.data[:, 0]
Y = housing.target

print(X.shape)
print(Y.shape)
```

# Defs

```python
w: float = np.random.randn()
b: float = np.random.randn()
```

```python
def weighted_sum(b, w):
    return b + X * w
```

```python
y_pred = weighted_sum(b, w)
y_pred.shape
```

```python
def mse(y_pred):
    return np.mean((y_pred - Y) ** 2)
```

# MSE derivative

```python
# true derivative
dmse_dyp = jacobian(mse)(y_pred)
dmse_dyp.shape
```

$$
\begin{align}
\frac{\partial \text{MSE}}{\partial \hat{y}_{p}} &=
\frac{\partial}{\partial \hat{y}_{p}} \left( 
	\frac{1}{m} \sum_{i=1}^{m} \left(\hat{y}_{i} - y_{i} \right)^{2}
\right) \\
&= \frac{1}{m} \sum_{i=1}^{m} \frac{\partial}{\partial \hat{y}_{p}}
\left( \left(\hat{y}_{i} - y_{i} \right)^{2} \right) \\
&= \frac{2}{m} \sum_{i=1}^{m} \left(\hat{y}_{i} - y_{i} \right)
\frac{\partial \hat{y}_{i}}{\partial \hat{y}_{p}} \\
&= \frac{2}{m} \sum_{i=1}^{m} \left(\hat{y}_{i} - y_{i} \right) \delta_{ip} \\
&= \frac{2}{m} \left(\hat{y}_{p} - y_{p} \right)
\end{align}
$$

for all $p = 1, \ldots, m$.

```python
# our derivative
our_dmse_dyp = 2 * (y_pred - Y) / len(Y)
our_dmse_dyp.shape
```

```python
mape(our_dmse_dyp, dmse_dyp)
```

# Weighted sum derivative

## respect to bias

```python
dyp_db = jacobian(weighted_sum, 0)(b, w)
dyp_db.shape
```

$$
\begin{align}
\frac{\partial \hat{y}_{p}}{\partial b} &=
\frac{\partial}{\partial b} \left(b + w x_{p} \right) \\
&= 1
\end{align}
$$

```python
our_dyp_db = np.ones_like(Y)
our_dyp_db.shape
```

```python
mape(our_dyp_db, dyp_db)
```

## respect to weight

```python
dyp_dw = jacobian(weighted_sum, 1)(b, w)
dyp_dw.shape
```

$$
\begin{align}
\frac{\partial \hat{y}_{p}}{\partial w} &= \frac{\partial}{\partial w}
\left( b + w x_{p} \right) \\
&= \frac{\partial}{\partial w} \left( w x_{p} \right) \\
&= x_{p}
\end{align}
$$

for all $p = 1, \ldots, m$.

```python
our_dyp_dw = X
our_dyp_dw.shape
```

```python
mape(our_dyp_dw, dyp_dw)
```

# complete derivative

## respect to bias

```python
dmse_db = jacobian(
    lambda b_: 
    mse(weighted_sum(b_, w))
, 0)(b)

dmse_db
```

$$
\begin{align}
\frac{\partial \text{MSE}}{\partial b} &= 
{\color{Cyan} \frac{\partial \text{MSE}}{\partial \hat{y}_{p}} }
{\color{Orange} \frac{\partial \hat{y}_{p}}{\partial b} } \\
&= {\color{Cyan} \frac{2}{m} \left(\hat{y}_{p} - y_{p} \right)}
{\color{Orange} 1_{p}} \\
&= \frac{2}{m} \left<\hat{\mathbf{y}} - \mathbf{y}, \mathbf{1} \right> \\
&= \frac{2}{m} \left( \hat{\mathbf{y}} - \mathbf{y} \right)^{\top} \mathbf{1}
\end{align}
$$

where $\mathbf{1} \in \mathbb{R}^{m}$.

```python
our_dmse_db = 2 / len(Y) * np.sum(y_pred - Y)
our_dmse_db
```

```python
mape(our_dmse_db.item(), dmse_db.item())
```

***

Using einsum

```python
mape(
    np.einsum('p,p->', our_dmse_dyp, our_dyp_db).item(),
    dmse_db.item()
)
```

## respect to weight

```python
dmse_dw = jacobian(
    lambda w_: 
    mse(weighted_sum(b, w_))
, 0)(w)

dmse_dw
```

$$
\begin{align}
\frac{\partial \text{MSE}}{\partial w} &= \
{\color{Cyan} \frac{\partial \text{MSE}}{\partial \hat{y}_{p}}}
{\color{Orange} \frac{\partial \hat{y}_{p}}{\partial w}} \\
&= {\color{Cyan} \frac{2}{m} \left(\hat{y}_{p} - y_{p} \right)}
{\color{Orange} x_{p}} \\
&= \frac{2}{m}  \left<\hat{\mathbf{y}} - \mathbf{y}, \mathbf{x} \right> \\
&= \frac{2}{m} \left( \hat{\mathbf{y}} - \mathbf{y} \right)^{\top}  \mathbf{x}
\end{align}
$$

```python
our_dmse_dw = 2 / len(Y) * np.dot(y_pred - Y, X)
our_dmse_dw
```

```python
mape(our_dmse_dw.item(), dmse_dw.item())
```

***

Using einsum

```python
mape(
    np.einsum('p,p->', our_dmse_dyp, our_dyp_dw).item(),
    dmse_dw.item()
)
```
